```python
df = pd.read_csv('datalog.csv')
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'].str.split('-').str[1]) - pd.Timedelta(hours=4,minutes=44)
```

```python
df['interval'] = 0
```

```python
df.head()
```

```python
df_1.head()
```

```python
df_2 = df[df['timestamp'] >= '11:30:00']
df_2 = df_2[df_2['timestamp'] < '12:00:00']
df_2['interval'] = 2
```

```python
df_3 = df[df['timestamp'] >= '12:00:00']
df_3 = df_3[df_3['timestamp'] < '12:30:00']
df_3['interval'] = 3
```
